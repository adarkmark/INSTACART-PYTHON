### Step 3: Create a 'region' column to compare spending habits with different regions

```python
# Create an empty container
region = []

for value in df_ords_prod_cust['State']:
    if value in ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey']:
        region.append('Northeast')
    elif value in ['Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']:
        region.append('Midwest')
    elif value in ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana']:
        region.append('South')
    else:
        region.append('West')
```

```python
region
```

```python
# Add 'region' column to df_ords_prod_cust dataframe

df_ords_prod_cust['region'] = region
```

```python

# Check value counts

df_ords_prod_cust['region'].value_counts(dropna = False)
```

```python
# Check output

df_ords_prod_cust.head()
```

```python
print(df_ords_prod_cust.columns)
```

```python
# Create a crosstab between 'region' and 'spending_flag'
region_spending = pd.crosstab(df_ords_prod_cust['region'], df_ords_prod_cust['spending_flag'], dropna = False)
```

## Step 4: Create an exclusion flag for low-activity customers (less than 5 orders)

```python
# Create a 'customer_activity' column

df_ords_prod_cust.loc[df_ords_prod_cust['max_order'] >= 5, 'customer_activity'] = 'High-activity'
```

```python
df_ords_prod_cust.loc[df_ords_prod_cust['max_order'] < 5, 'customer_activity'] = 'Low-activity'
```

```python
# Check values count

df_ords_prod_cust['customer_activity'].value_counts(dropna = False)
```

```python
# Create a subset with only high-activity customers

df_high_activity_cust = df_ords_prod_cust[df_ords_prod_cust['customer_activity'] == 'High-activity']
```
